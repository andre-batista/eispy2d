# Case Study - Usage Example

After runing the script `example_case_study.py`, a file is saved with the results. Then, we can load it and use the class routines to analyze the results.

```python
import casestudy as cst
mystudy = cst.CaseStudy('void')
mystudy.importdata('cst_basic')
```

First, we want to visualize the best reconstructions accomplished by each algorithm according to the $\zeta_{\epsilon PAD}$ indicator:

```python
import result as rst
mystudy.reconstruction(mode=cst.BEST_EXECUTION,
                       indicator=rst.REL_PERMITTIVITY_PAD_ERROR,
                       include_true=True,
                       show=True)
```

Now, we want to check the convergence of the objective function for each algorithm:

```python
mystudy.convergence(rst.OBJECTIVE_FUNCTION,
                    sample_rate=20,
                    mean=True,
                    show=True,
                    fontsize=15,
                    method=None,
                    widths=10,
                    color='tab:blue',
                    title=False,
                    yscale='log')
```

Next, we want to look at the $\zeta_{\epsilon PAD}$ value reached by each algorithm considering all the executions:

```python
mystudy.boxplot('zeta_epad',
                show=True,
                fontsize=15)
```

Now, we want to compare the average performance of each algorithm considering the $\zeta_{\epsilon PAD}$ indicator:

```python
mystudy.compare('zeta_epad', all2all=True)
```

The confidence interval displayed above is calculated considering logarithmic transformation. But, we want to see the confidence interval of means for the original data:

```python
mystudy.confint('zeta_epad', show=True, fontsize=15)
```
